```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movies_df = pd.read_csv('/home/dans/Downloads/pyp/py2/movies.csv')
movies_df
```

```python
#columns i need
impColumns = ['genres','title','keywords','cast','director', 'tagline']
```

```python
#replacing null values with null strings
for column in impColumns:
    movies_df[column] = movies_df[column].fillna('')
```

```python
selectColumns = movies_df['title']+""+movies_df['genres']+""+movies_df['cast']+""+movies_df['director']+''+movies_df['tagline']

selectColumns
```

```python
#vectorizer 
vectorizer = TfidfVectorizer()

vectorizedColumns = vectorizer.fit_transform(selectColumns)
print(vectorizedColumns)
```

```python
#cosine similarity scores of the data
similarity = cosine_similarity(vectorizedColumns)
similarity.shape
```

```python
#getting user movie choice
userMovie = input("enter movie name:")
```

```python
#create a list of all the movies in the dataste 
moviesList = movies_df['title'].tolist()
```

```python
# compare the movie name provided by the user to one in the dataset to find the closest match using difflib
getCloseMatch = difflib.get_close_matches(userMovie, moviesList)
matchedMovies = getCloseMatch[0]
matchedMovies
```

```python
#Find the index of the matched movie on the dataset
matchedMovieIndex = movies_df[movies_df.title == matchedMovies]['index'].values[0]
matchedMovieIndex
```

```python
# Get a list of similar movies using the cosine similarity and the index of the matched movies
similarity_score = list(enumerate(similarity[matchedMovieIndex]))
```

```python
#sort the movies from highest  similarity score to lowest
sortedMovies = sorted(similarity_score, key = lambda x:x[1], reverse = True)
```
